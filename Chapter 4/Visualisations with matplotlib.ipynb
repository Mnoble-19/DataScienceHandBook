# Visualisation with matplotlib

## Intro

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
```

```python
%matplotlib inline
x = np.linspace(0,10,100)

fig = plt.figure()
plt.plot(x, np.sin(x), '*')
plt.plot(x, np.cos(x), "-")
```

```python
fig.savefig('first_viz.png') #saves in that specific directory
```

## Demonstrating object-oriented interface

```python
plt.figure()

#Creates 1st panels
plt.subplot(2,1,1) # (rows,colums, panel no..)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

```

```python
## Better way to do it

fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## Simple Line Plots
* For all Matplotlib plots, we start by creating a figure and an axes.

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```
