# Visualisation with matplotlib

## Intro

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
```

```python
% matplotlib inline
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '*')
plt.plot(x, np.cos(x), "-")
```

```python
fig.savefig('first_viz.png')  #saves in that specific directory
```

## Demonstrating object-oriented interface

```python
plt.figure()

#Creates 1st panels
plt.subplot(2, 1, 1)  # (rows,colums, panel no..)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

```

```python
## Better way to do it

fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## Simple Line Plots
* For all Matplotlib plots, we start by creating a figure and an axes.
* plt.xlabel() → ax.set_xlabel()
* plt.ylabel() → ax.set_ylabel()
* plt.xlim() → ax.set_xlim()
* plt.ylim() → ax.set_ylim()
* plt.title() → ax.set_title()


```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue', linestyle='-.')
plt.plot(x, np.sin(x - 1), color='g', linestyle='-')
plt.plot(x, np.sin(x - 2), color='0.75')
plt.plot(x, np.sin(x - 3), color='#FFDD44')
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
```

```python
##Changing limits
plt.plot(x, np.sin(x), '-g', label="sin(x)")
plt.plot(x, np.cos(x), ':.r', label="cos(x)")
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.axis([-1, 11, -1.5, 1.5])
plt.title('Noble\'s first plot')
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.legend()
```

## Scatter plots

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'og')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label=f'marker={marker}')
    plt.legend(numpoints=1)
    plt.xlim(0,1.0)
```

```python
plt.scatter(x,y,marker='o')
```

The primary difference of plt.scatter from plt.plot is that it can be used to create scatter plots where the properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng. randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
rng = np.random.RandomState(0)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
                       s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
