# Visualisation with matplotlib

## Intro

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
```

```python
% matplotlib inline
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '*')
plt.plot(x, np.cos(x), "-")
```

```python
fig.savefig('first_viz.png')  #saves in that specific directory
```

## Demonstrating object-oriented interface

```python
plt.figure()

#Creates 1st panels
plt.subplot(2, 1, 1)  # (rows,colums, panel no..)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

```

```python
## Better way to do it

fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## Simple Line Plots
* For all Matplotlib plots, we start by creating a figure and an axes.
* plt.xlabel()  ax.set_xlabel()
* plt.ylabel()  ax.set_ylabel()
* plt.xlim()  ax.set_xlim()
* plt.ylim()  ax.set_ylim()
* plt.title()  ax.set_title()


```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x - 0), color='blue', linestyle='-.')
plt.plot(x, np.sin(x - 1), color='g', linestyle='-')
plt.plot(x, np.sin(x - 2), color='0.75')
plt.plot(x, np.sin(x - 3), color='#FFDD44')
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
```

```python
##Changing limits
plt.plot(x, np.sin(x), '-g', label="sin(x)")
plt.plot(x, np.cos(x), ':.r', label="cos(x)")
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.axis([-1, 11, -1.5, 1.5])
plt.title('Noble\'s first plot')
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.legend()
```
